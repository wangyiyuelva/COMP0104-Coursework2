```python
import pandas as pd
import matplotlib.pyplot as plt

project_list = [
    'zeppelin',
    'parquet-mr',
    'ignite',
]
```

```python
for repo_name in project_list:
    csv_file = f'./analyseCommitSize/{repo_name}-table_commit_size.csv'
    df = pd.read_csv(csv_file)

    # First Figure
    fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 5))
    ax[0].scatter(df['files'], df['num_test_files'], alpha=0.3)
    ax[0].set_xlabel('Number of Changed Files')
    ax[0].set_ylabel('Number of Test files')
    ax[0].set_title(f'{repo_name} \n Correlation between Files and Inclusion of Test Files')

    # Second Figure
    ax[1].scatter(df['lines'], df['num_test_files'], alpha=0.3)
    ax[1].set_xlabel('Number of Changed Lines')
    ax[1].set_ylabel('Number of Test files')
    ax[1].set_title(f'{repo_name} \n Correlation between Lines and Inclusion of Test Files')

    plt.tight_layout()

    plt.show()
```
